```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
x_train = sc_X.fit_transform(x_train)
x_test = sc_X.transform(x_test)
sc_y = StandardScaler()
y_train = sc_y.fit_transform(y_train.values.shape(-1,1))
y_test = sc_y.transform(y_test.values.reshape(-1,1))
```

```python
#create and fit the model
model = LinearRegression()
model.fit(x_train,y_train) # Input should be list of a list, so we should use values.reshape(-1,1)
print(y_test)
y_tested=model.predict(x_test)
print(y_tested)
print(model.score(x_test,y_test))
```
